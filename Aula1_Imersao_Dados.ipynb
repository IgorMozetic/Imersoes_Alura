##Primeiro passo - Importar a base de dados

```python
import pandas as pd #importando a biblioteca que irá ser utilizada e dando um apelido a ela

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true' #passando a url para a variável 'url_dados'

df_dados = pd.read_csv(url_dados, compression = 'zip') #lendo a base de dados dentro de um arquivo zip e passando para a variàvel 'df_dados'
display(df_dados) #exibindo a base de dados
```

##Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

```python
df_dados['tratamento'].unique()
```

```python
df_dados['tratamento'].value_counts(normalize = True)
```

Como podemos perceber, existe uma grande diferença entre os tratamentos com drogas e com controle, portanto, vamos ver quais as drogas os tratamentos utilizam

```python
dados_com_controle.shape[0]
```

```python
dados_com_droga.shape[0]
```

##Desafio 05: Procurar na documentação o método query(pandas).

O método `query()` se encontra dentro da biblioteca do pandas e é muito útil para realizar consultas em tabelas/Dataframes(df) utilizando uma expressão booleana, ou seja, se a expressão for True, faça isso.

Ele pode ser executado como nesse exemplo a seguir: 

`DataFrame.query( expr , inplace = False , ** kwargs )[fonte]`

####Outro exemplo

df = pd.DataFrame({
  
                   'A': range(1, 6),
                   
                   'B': range(10, 0, -2),

                   'C': range(10, 5, -1)})

df

      A   B    C

      0  1  10   10

      1  2   8    9

      2  3   6    8

      3  4   4    7

      4  5   2    6

df.query('A > B')

      A  B  C

      4  5  2  6

##Desafio 06: Renomear as colunas tirando o hífen.

```python
colunas = df_dados.rename(lambda i: i.replace('-', ''), axis=1)
display(colunas)
```

O comando `.rename()` alterou os nomes das colunas, retirando o '-' e colocando um espaço nulo no lugar

O `axis=1` refere-se as colunas, axis = eixo, então `axis=0` é horizontal

##Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt
df_dados['tempo'].value_counts().plot.bar(color='#45b1b7', fontsize=10, figsize=(10,5))
plt.title('\nTempo gasto no Tratamento\n', family='monospace', color='#262626', fontsize=30)
plt.xlabel('\nHoras', color='#757575', fontsize=20)
plt.ylabel("Quantidade\n", color='#757575', fontsize=20)
```
